## Step 1 : Read the Train dataset

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

## Step 2 : Perform basic data quality checks

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep="first")
df = df.reset_index(drop=True)
df
```

## Step 3 : Seperate X and Y(species)

```python
X = df.drop(columns=["species"])
Y = df[["species"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts()
```

## Step 4 : Apply Preprocessing for X

```python
X.dtypes
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer # replaces missing value
from sklearn.preprocessing import StandardScaler  # Standardize the data values into standard format through fit_transform
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

## Step 5 - Train Test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    X_pre, Y, test_size=0.33, random_state=42
)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Step 6 : Model Building

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(
    max_depth=1,
    min_samples_split=5,
    min_samples_leaf=5,
    criterion="entropy"
)
dtc.fit(xtrain, ytrain)
```

```python
dtc.score(xtrain, ytrain)
```

```python
dtc.score(xtest, ytest)
```

## Hyper paramater tuning to improve model performance

```python
params = {
    "max_depth": [2, 3, 4, 5, 6],
    "min_samples_split": [5, 6, 7, 8, 9],
    "min_samples_leaf": [5, 6, 7, 8, 9],
    "criterion": ["gini", "entropy"]
}
```

```python
from sklearn.model_selection import GridSearchCV
model = DecisionTreeClassifier()
gscv = GridSearchCV(
    estimator= model,
    param_grid= params,
    scoring= "f1_macro",
    cv = 5
)
gscv.fit(xtrain, ytrain)
```

```python
gscv.best_params_   # values may vary from others as when decision tree is built there is randomness to it
```

```python
gscv.best_score_          # gives f1 macro, this value is always closer to what others get
```

```python
best_dtc = gscv.best_estimator_
best_dtc
```

## Step 7 : Model Evaluation

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc, xtrain, ytrain)
```

```python
ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```

```python
ypred_test = best_dtc.predict(xtest)
```

```python
ypred_test[0:5]
```

```python
ytest.head()  # for comparison
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```
