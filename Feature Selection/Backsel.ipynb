## Step 1 - Read the dataset

### Step 3 - Seperate X and Y(Weight) values
id should be dropped because stastically insignificant

```python
X = df.drop(columns = ["id", "Weight"])
Y = df[["Weight"]]
```

```python
X.head()
```

```python
Y.head()
```

### Step 4 - Feature selection Pipeline
Categorical - Ordinal(lable) encoding

```python
cat = list(X.columns[X.dtypes == "object"])
con = list(X.columns[X.dtypes != "object"])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = make_pipeline(
    SimpleImputer(strategy = "mean"),
    StandardScaler()
)
```

```python
cat_pipe1 = make_pipeline(
    SimpleImputer(strategy = "most_frequent"),   # if written "mode" it wont show error here but when recalled
    OrdinalEncoder()
)
```

```python
pre1 = ColumnTransformer([
    ("num", num_pipe1, con),
    ("cat", cat_pipe1, cat)
]).set_output(transform="pandas")
```

```python
X_pre = pre1.fit_transform(X)
# this assigns numbers to alphabetical data and now can be passed feature selection pipeline
X_pre.head()
```

```python
X_pre.isna().sum()
```

```python
pre1
```

### Step 5 - Forward feature Selection (Automated or manual if required)

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sel = SequentialFeatureSelector(
    LinearRegression(), direction="backward", n_features_to_select="auto")
sel.fit(X_pre, Y)
imp_cols = sel.get_feature_names_out()
imp_cols
```

```python
len(imp_cols)
```

### in above example we tried it with n_features_to_select=10 but when auto option was selected machine stopped at 13 means peak value was achieved at 13 features (columns) i.e a point where bias and variance coinside (highest R2 score) and error is least


### To get original column names 

```python
imp_cols[0]
```

```python
imp_cols[0].split("__")
```

```python
imp_cols[0].split("__")[1]
```

```python
sel_cols = [col.split("__")[1] for col in imp_cols]
sel_cols
```

```python
X_sel = X[sel_cols]
X_sel.head()
```

### Step 6 - Create a final pipeline
cat - OneHotEncoder

```python
cat_sel = list(X_sel.columns[X_sel.dtypes == "object"])
con_sel = list(X_sel.columns[X_sel.dtypes != "object"])
```

```python
con_sel
```

```python
cat_sel
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = make_pipeline(
    SimpleImputer(strategy="mean"),
    StandardScaler()
)
```

```python
cat_pipe2 = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False)   # o/p required in array or dataframe and not sparse
)
```

```python
pre2 = ColumnTransformer([
    ("num", num_pipe2, con_sel),
    ("cat", cat_pipe2, cat_sel)
]).set_output(transform="pandas")
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre.head()
```

```python
pre2
```

### Step 7 - Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    X_sel_pre, Y, test_size=0.2, random_state=21
)
```

```python
ytrain.head()
```

### Step 8 - Model Building

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, \
                            mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)
    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"MSE : {mse:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.4f}")
    print(f"R2 : {r2:.4f}")
```

```python
evaluate_model(model, xtrain, ytrain)
```

```python
evaluate_model(model, xtest, ytest)
```
