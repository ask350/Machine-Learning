## Step 3 - Seperate X and Y(booking_status)

```python
X = df.drop(columns=["id", "booking_status"])
Y = df[["booking_status"]]
```

## Step 4 - Apply preprocessing for X

```python
X.dtypes
```

### as  there are only numeric values we need single numeric pipeline

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

## Step 5 - Train-Test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

## Step 6 -Model Bulding

```python
import xgboost as xgb
```

```python
model = xgb.XGBClassifier(
    n_estimators = 200,
    max_depth = 3
)
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

## No Hyperparameter tuning required

## Step 7 - Evaluate model

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

```python
ypred_test = model.predict(xtest)
ypred_test[0:5]
```

```python
ytest.head()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

## Step 8 - Out of Sample Prediction

```python
xnew = pd.read_csv("test_booking.csv")
xnew.head()
```

```python
xnew.shape
```

```python
xnew2 = xnew.drop(columns=["id"])
xnew2.head()
```

```python
num_pipe
```

```python
xnew2_pre = num_pipe.transform(xnew2)
xnew2_pre.head()
```

```python
preds = model.predict(xnew2_pre)
preds[0:5]
```

```python
model.classes_
```

```python
probs = model.predict_proba(xnew2_pre)
probs
```

```python
probs = model.predict_proba(xnew2_pre)[:, 1]      # only interested in cancellation probability
probs[0:5]
```

```python
results = xnew[["id"]]
results["booking_pred"] = preds
results["cancel_prob"] = probs
```

```python
results = results.round(4)
results
```

```python
results["booking_pred"].value_counts()
```

```python
results["booking_pred"].value_counts().plot(kind="bar")
```

```python
results.to_csv("XGBResults.csv", index=False)
```
